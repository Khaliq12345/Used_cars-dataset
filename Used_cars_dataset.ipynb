```python
n = []
pri = []
k = []
p = []
d = []
a = []

for i in range(0,13013,13):
    print(i)
    url = f'https://www.carsguide.com.au/buy-a-car/all-used/qld/all-locations/all-bodytypes/all-makes?distanceFromMe=Brisbane%2CQLD&searchOffset={i}&searchLimit=13'
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'lxml')
    #scraping data
    boxs = soup.find_all('div', attrs={'class':'carListing--content'})
    for box in boxs:
        try:
            name = box.find('h5', attrs={'class':'carListing--textHeading'}).text
            n.append(name)
        except:
            n.append(np.NAN)
            
        try:
            price = box.find('h6', attrs={'class':'carListing--textPrice'}).text
            pri.append(price)
        except:
            pri.append(np.NAN)
            
        try:
            km = box.find('li',attrs={'class':'icon icon-gauge'}).text
            k.append(km)
        except:
            k.append(np.NAN)
            
        try:
            process = box.find('li',attrs={'class':'icon icon-gears'}).text
            p.append(process)
        except:
            p.append(np.NAN)
            
        try:
            dealer = box.find('li',attrs={'class':'icon sellerCircle dealer'}).text
            d.append(dealer)
        except:
            d.append(np.NAN)
            
        try:
            address = box.find('li',attrs={'class':'icon icon-pin'}).text
            a.append(address)
        except:
            a.append(np.NAN)
                
data = pd.DataFrame({'Name':n, 'Price':pri, 'Kilometer':k, 'Process':p, 'Dealer':d, 'Address':a})
data.to_csv('used_cars_dataset.csv')
data
```
